## Data Preprocessing

### One-Hot Encoding of Categorical Variables

One-hot encoding is performed on the categorical variables such as 'Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status', 'Income_Category', and 'Card_Category' to convert them into a format that can be provided to machine learning algorithms to enable better model training.
```python
# Perform one-hot encoding on the specified categorical variables
categorical_columns = ['Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']
df_encoded = pd.get_dummies(df, columns=categorical_columns)

# Display the shape of the new dataframe and the first few rows to confirm the encoding
df_encoded_shape = df_encoded.shape
df_encoded_head = df_encoded.head()

df_encoded_shape, df_encoded_head
```

## Removing Constant & Duplicate Features 

In the data preprocessing phase, constant features were identified and removed. Constant features, which have the same value for all observations, do not contribute any information that can be used for prediction and can be eliminated to reduce model complexity and computational cost.

Additionally, duplicate features were identified and removed. Duplicate features are those that are identical to others across all observations and therefore redundant. Removing duplicates helps in reducing the dimensionality of the dataset, ensuring that each feature provides unique information to the model and improving the efficiency of model training.

```python
# Identifying constant features
constant_features = [col for col in df.columns if df[col].nunique() == 1]

# Dropping constant features
df.drop(columns=constant_features, inplace=True)

# Identifying duplicate features
duplicate_features = df.columns[df.T.duplicated()]

# Dropping duplicate features
df.drop(columns=duplicate_features, inplace=True)

# Display the removed constant and duplicate features
constant_features, duplicate_features
```

## Filter Methods - Statistical Significance Testing

### Mutual Information Method

Mutual Information (MI) has been utilized to quantify the amount of information each feature shares with the target variable, `Attrition_Flag_Attrited Customer`. This method helps in identifying features that have the strongest relationships with the target variable. Features with higher MI scores are more predictive and thus are likely more important for our model. This non-linear dependency evaluation is crucial for understanding the underlying patterns in the data that may not be captured by linear methods such as correlation. The MI scores have been sorted in descending order to prioritize the features that could contribute most to the model's predictive power.

```python
from sklearn.feature_selection import mutual_info_classif

# Prepare the features and target variable for mutual information
X = df_final.drop(['Attrition_Flag_Attrited Customer', 'Attrition_Flag_Existing Customer'], axis=1)
y = df_final['Attrition_Flag_Attrited Customer']

# Calculate Mutual Information
mi_scores = mutual_info_classif(X, y, random_state=0)

# Create a Series for better visualization
mi_scores_series = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)

mi_scores_series
```

### Chi-Squared Test

The Chi-Squared Test is utilized here to evaluate the independence of binary features derived from one-hot encoding with respect to the target variable, `Attrition_Flag_Attrited Customer`. Although our dataset is preprocessed with all features encoded and scaled, the Chi-Squared Test is still applicable because it requires categorical input and can be used on binary features. The test provides Chi2 scores and p-values for each feature, with a lower p-value indicating a feature that is likely to be more important and statistically significant with respect to the target variable. Features with higher Chi2 scores and lower p-values are prioritized as they are more likely to have a meaningful association with the target, helping in the selection of relevant features for the predictive model.


```python
from sklearn.feature_selection import chi2

# Since our dataset is already encoded and scaled, we directly select binary features for the Chi-squared test.
# For this example, we'll consider all features as they are encoded and treat them for the Chi-squared test.
# Note: Normally, Chi-squared is applied before encoding and scaling, primarily on raw categorical data.

# Filter for binary features only (resulting from one-hot encoding)
binary_features = [col for col in X.columns if df_final[col].nunique() == 2]

# Select only binary features for this test
X_binary = X[binary_features]

# Apply the Chi-squared test
chi_scores, chi_p_values = chi2(X_binary, y)

# Create a DataFrame for better visualization
chi_results = pd.DataFrame({'Feature': binary_features, 'Chi2 Score': chi_scores, 'P-value': chi_p_values}).sort_values(by='Chi2 Score', ascending=False)

chi_results
```

## Feature Selection

Feature selection has been performed using three different methods to identify the most informative features for our predictive model:

1. **Random Forest Importance:** This method leverages the feature importance attribute provided by Random Forest models. By evaluating the contribution of each feature to the model's performance, we can rank the features based on their importance and select the top features for further analysis.

2. **Recursive Feature Elimination (RFE):** RFE is an iterative feature selection technique that recursively removes features from the dataset and evaluates the model's performance after each removal. By selecting features that contribute the most to the model's accuracy, RFE helps in identifying the subset of features that are most relevant for prediction.

3. **XGBoost Feature Importance:** XGBoost is an ensemble learning algorithm known for its effectiveness in feature selection. Similar to Random Forest, XGBoost provides a feature importance score based on the contribution of each feature to the model's performance. By ranking the features according to their importance, we can identify the key features that have the most significant impact on the predictive power of the model.


## Random Forest

```python
# Re-import necessary libraries and load the dataset, then proceed with feature importance analysis
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)

# Initialize the RandomForest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the classifier to the training data
rf_classifier.fit(X_train, y_train)

# Obtain feature importance scores from the fitted RandomForest model
feature_importances_rf = rf_classifier.feature_importances_

# Create a DataFrame to display feature importances
feature_importances_df = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': feature_importances_rf
}).sort_values(by='Importance', ascending=False).reset_index(drop=True)

# Perform permutation importance on the test set to get a model-agnostic view of feature importance
perm_importance_result = permutation_importance(rf_classifier, X_test, y_test, n_repeats=10, random_state=42)

# Create a DataFrame to display permutation importance results
perm_importance_df = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': perm_importance_result.importances_mean
}).sort_values(by='Importance', ascending=False).reset_index(drop=True)

# Display the top features based on RandomForest feature importance
feature_importances_df.head(), perm_importance_df.head()
```

## Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE

# Assuming X_encoded_reduced and y are already defined

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded_reduced, y, test_size=0.3, random_state=42)

# Initialize the model to be used for feature selection
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Initialize RFE with the model and desired number of features
rfe = RFE(estimator=model, n_features_to_select=10)

# Fit RFE
rfe.fit(X_train, y_train)

# Transform the data according to the RFE
X_train_rfe = rfe.transform(X_train)
X_test_rfe = rfe.transform(X_test)

# Fit the model on the transformed data
model.fit(X_train_rfe, y_train)

# Evaluate the model
print("Model score with selected features:", model.score(X_test_rfe, y_test))

# List of selected features
selected_features = X_encoded_reduced.columns[rfe.support_]
print("Selected features:", selected_features)
```

## Dimensionality Reduction

As part of the preprocessing pipeline, Principal Component Analysis (PCA) has been applied for dimensionality reduction. PCA is a widely used technique for reducing the number of features in high-dimensional datasets while preserving most of the variance in the data.

The reduced-dimensional representation obtained through PCA can help alleviate issues associated with high-dimensional data, such as the curse of dimensionality, computational complexity, and overfitting. By retaining the most important information while discarding redundant or less informative features, PCA facilitates more efficient model training and interpretation.

In this notebook, PCA has been employed to reduce the dimensionality of the feature space, with the number of principal components chosen based on the desired level of variance retention or reduction in dimensionality. This preprocessing step aims to streamline the modeling process and improve the performance and interpretability of the predictive model.

## Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming X_encoded_reduced is our DataFrame after encoding and removing correlated features
# And that y is the target variable

# Standardize the features before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded_reduced)

# Initialize PCA, here we choose enough components to explain 95% of the variance
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame of the PCA results
X_pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])

# Display the shape of the original data and the PCA transformed data
print(f'Original shape: {X_encoded_reduced.shape}')
print(f'Transformed shape: {X_pca_df.shape}')

# Display the DataFrame with PCA components
X_pca_df.head()
```

## Alternative Feature Selection Methods



#### 1. Regularization Methods
- **Lasso (L1 Regularization):** Shrinks coefficients to zero, effectively performing feature selection.
- **Ridge (L2 Regularization):** Encourages smaller, more diffuse coefficients.
- **Elastic Net:** Combines L1 and L2 regularization to select features.

#### 2. Model-Based Selection
- **Random Forest Importance:** Leverages the feature importance from Random Forests.
- **Gradient Boosting Machines (GBM):** Utilizes GBM feature importance for feature selection.

#### 3. Stability Selection
- **Randomized Lasso/Logistic Regression:** Applies regularization over subsamples to select features.

#### 4. Recursive Feature Addition or Elimination
- **Recursive Feature Addition (RFA):** Begins with no features and adds the most improving ones.
- **Recursive Feature Elimination (RFE):** Starts with all features, removing the least important iteratively.

#### 5. Dimensionality Reduction Techniques
- **Principal Component Analysis (PCA):** Transforms features into principal components.
- **t-SNE:** Maps high-dimensional data to a lower-dimensional space.
- **Autoencoders:** Uses neural networks for unsupervised feature extraction.

#### 6. Genetic Algorithms
- **Evolutionary Feature Selection:** Employs evolutionary strategies to select optimal features.

#### 7. Feature Shuffling
- **Permutation Importance:** Assesses importance by the increase in prediction error when a feature's values are shuffled.

#### 8. Bayesian Model Averaging
- **Bayesian Approaches:** Incorporates model uncertainty into the feature selection process.

#### 9. Information-Theoretic Methods
- **Minimum Redundancy Maximum Relevance (mRMR):** Chooses features with high relevance and low redundancy.

#### 10. Biclustering
- **Biclustering Algorithms:** Finds clusters of features and samples for feature selection.
